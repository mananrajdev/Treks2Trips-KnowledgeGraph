```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException,StaleElementReferenceException
from bs4 import BeautifulSoup
import re
import sys
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/usr/local/bin/chromedriver", chrome_options=options)
```

```python
driver = webdriver.Chrome("/usr/local/bin/chromedriver", chrome_options=options) #watch out, change if you are not using Chrome
driver.get('https://www.alltrails.com/us/states')

numstates = 0
while True:
    try:
        showmore=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH ,'/html/body/div[3]/div[3]/div/div[3]/div/div[3]/div/button')))
        showmore.click()
        numstates = int(driver.find_element_by_class_name('styles-module__showing___14AKF').text.split('- ')[1].split('of')[0])
    except TimeoutException:
        break
    except:
        break
        
print(numstates)

states_page_source = driver.page_source  
```

```python
states = {}
for idx in range(50):
    path = '/html/body/div[3]/div[3]/div/div[3]/div/div[2]/div[2]/div[' + str(idx+1) +']/a'
    state_selector = driver.find_element_by_xpath(path)
    title = state_selector.get_attribute('title')
    link = state_selector.get_attribute('href')
    states[title] = link
    
print(len(states))
print(states)
```

```python
final_treks = {}
done_state = []
```

```python
def driver_code(state_url):
    driver = webdriver.Chrome("/usr/local/bin/chromedriver", chrome_options=options) #watch out, change if you are not using Chrome
    driver.get(state_url)

    numtreks = 0
    while True:
        try:
            showmore=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH ,'/html/body/div[3]/div[3]/div/div[3]/div[2]/div[2]/div/button')))
            showmore.click()
            numtreks = int(driver.find_element_by_xpath('/html/body/div[3]/div[3]/div/div[3]/div[2]/div[2]/div/div').text.split('- ')[1].split('of')[0])
        except TimeoutException:
            break
        except:
            break

    print('Treks in the state:', numtreks)

    page_source = driver.page_source
    return driver, page_source, numtreks
```

```python
def treks_by_state(state, driver, page_source, numtreks):
    
    if numtreks > 1000:
        r = 1000
    else:
        r = numtreks
 
    for idx in range(r):
        # Link
        path = '/html/body/div[3]/div[3]/div/div[3]/div[2]/div[1]/div[2]/div[' + str(idx+1) +']/a'
        treks_selector = driver.find_element_by_xpath(path)
        title = treks_selector.get_attribute('title')
        link = treks_selector.get_attribute('href')
        
        # Time
        path = '/html/body/div[3]/div[3]/div/div[3]/div[2]/div[1]/div[2]/div[' + str(idx+1) + ']/div[2]/div[3]/span[3]'
        try:
            time = driver.find_element_by_xpath(path).text.split('Est. ')[1]
        except:
            time = 'None'
        final_treks[title] = tuple((state, link, time))

    print('Treks Collected Yet:',len(final_treks))
```

```python
for state, url in states.items():
    if state not in done_state:
        print(url)
        driver, page_source, numtreks = driver_code(url)
        treks_by_state(state, driver, page_source, numtreks)
        done_state.append(state)
```

```python
def trek_driver(name, trek_url, state, time):
    driver = webdriver.Chrome("/usr/local/bin/chromedriver", chrome_options=options) #watch out, change if you are not using Chrome
    driver.get(trek_url)
    page_source = driver.page_source
    
    
#     difficulty = driver.find_element_by_class_name('styles-module__diff___22Qtv styles-module__easy___bPX-K styles-module__selected___3fawg').text
#     park = driver.find_element_by_class_name('xlate-none styles-module__location___11FHK styles-module__location___3wEnO').text
        
#     /html/body/div[3]/div[4]/div/div/div[1]/div/div/div/div[1]/div/div/span[1]
    trek_items = []
    try:
        dif_path = '/html/body/div[3]/div[3]/div/div/div[1]/div/div/div/div[1]/div/div/span[1]'
        difficulty = driver.find_element_by_xpath(dif_path).text
    except:
        return None
#         dif_path = '/html/body/div[3]/div[4]/div/div/div[1]/div/div/div/div[1]/div/div/span[1]'
#         difficulty = driver.find_element_by_xpath(dif_path).text
          
    
    try:
        park_path = '/html/body/div[3]/div[3]/div/div/div[1]/div/div/div/div[1]/div/a'
        park = driver.find_element_by_xpath(park_path).text
    except:
        return None
#         park_path = '/html/body/div[3]/div[4]/div/div/div[1]/div/div/div/div[1]/div/a'
#         park = driver.find_element_by_xpath(park_path).text
        
        
    try:
        len_path = '/html/body/div[3]/div[3]/div/div/div[2]/div[1]/article/section[2]/div/span[1]/span[2]'
        length = driver.find_element_by_xpath(len_path).text
    except:
        return None
        len_path = '/html/body/div[3]/div[4]/div/div/div[2]/div[1]/article/section[2]/div/span[1]/span[2]'
        length = driver.find_element_by_xpath(len_path).text
        
    try:
        eg_path = '/html/body/div[3]/div[3]/div/div/div[2]/div[1]/article/section[2]/div/span[2]/span[2]'
        elevation_gain = driver.find_element_by_xpath(eg_path).text
    except:
        return None
#         eg_path = '/html/body/div[3]/div[4]/div/div/div[2]/div[1]/article/section[2]/div/span[2]/span[2]'
#         elevation_gain = driver.find_element_by_xpath(eg_path).text
        
    try:
        rt_path = '/html/body/div[3]/div[3]/div/div/div[2]/div[1]/article/section[2]/div/span[3]/span[2]'
        route_type = driver.find_element_by_xpath(rt_path).text
    except:
        return None
#         rt_path = '/html/body/div[3]/div[4]/div/div/div[2]/div[1]/article/section[2]/div/span[3]/span[2]'
#         route_type = driver.find_element_by_xpath(rt_path).text
        
    try:
        rs_path = '/html/body/div[3]/div[3]/div/div/div[1]/div/div/div/div[1]/div/div/span[2]/meta[1]'
        review_stars = driver.find_element_by_xpath(rs_path).get_attribute('content')
    except:
        return None
#         rs_path = '/html/body/div[3]/div[4]/div/div/div[1]/div/div/div/div[1]/div/div/span[2]/meta[1]'
#         review_stars = driver.find_element_by_xpath(rs_path).get_attribute('content')

    lst = driver.find_elements_by_class_name('tag') #big rounded active
    
    tags = []
    for idx, item in enumerate(lst):
#         print(idx, item)
#         print(item.get_attribute("innerHTML"))
#         path = '/html/body/div[3]/div[3]/div/div/div[2]/div[1]/article/section[3]/span[' + str(idx+1) + ']/span'
        tag = item.get_attribute('innerText') #'innerHTML'
        tags.append(tag)
    
    trek_items.append(name)
    trek_items.append(state)
    trek_items.append(difficulty)
    trek_items.append(park)
    trek_items.append(length)
    trek_items.append(elevation_gain)
    trek_items.append(route_type)
    trek_items.append(review_stars)
    trek_items.append(time)
    trek_items.append(tags)
    
    return trek_items
```

```python
treks = pd.read_csv('trek_url_list_info.csv', header=None)

for row, trek in treks.iterrows():
    if row not in done_trek:
        trek_items = trek_driver(trek[0], trek[3], trek[1], trek[2])
        if trek_items == None:
            not_done.append(row)
            print('Row Not Finished:', row)
        else:
            final_info[trek[0]] = trek_items
            done_trek.append(row)
            print('Row Finished:', row)
        
print(len(final_info))
```
